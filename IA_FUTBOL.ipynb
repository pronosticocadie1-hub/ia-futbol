```python

import pandas as pd

# Convertir los partidos a lista simple
lista_partidos = []

for p in partidos:
    try:
        lista_partidos.append({
            "local": p["homeTeam"]["name"],
            "visitante": p["awayTeam"]["name"]
        })
    except:
        pass

# Convertir a tabla
df = pd.DataFrame(lista_partidos)

# Guardar en archivo
df.to_csv("partidos.csv", index=False)

print("✅ Datos guardados en partidos.csv")
```
